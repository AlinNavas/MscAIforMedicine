```python
import os
import random
import shutil

for x in ['chroma_imgs', 'mel_spectrogram_imgs', 'mfcc_imgs', 'stft_imgs','spec_imgs']:

    # Set the path to the dataset folder
    DATASET_PATH = './data_pre/{type}'.format(type=x)

    # Set the test-train split ratio
    TRAIN_RATIO = 0.70
    VAL_RATIO = 0.15

    # Iterate over each subfolder
    for subfolder in os.listdir(DATASET_PATH):
        subfolder_path = os.path.join(DATASET_PATH, subfolder)
        # Get the list of files in the subfolder
        files = os.listdir(subfolder_path)
        
        # Create train, val, and test folders inside the subfolder
        train_folder = os.path.join(DATASET_PATH, 'train')
        val_folder = os.path.join(DATASET_PATH, 'val')
        test_folder = os.path.join(DATASET_PATH, 'test')
        os.makedirs(train_folder, exist_ok=True)
        os.makedirs(val_folder, exist_ok=True)
        os.makedirs(test_folder, exist_ok=True)
        
        # Shuffle the files randomly
        random.shuffle(files, random.seed(42))
        
        # Calculate the number of files for training, validation, and testing
        num_train_files = int(len(files) * TRAIN_RATIO)
        num_val_files = int(len(files) * VAL_RATIO)
        train_files = files[:num_train_files]
        val_files = files[num_train_files:num_train_files+num_val_files]
        test_files = files[num_train_files+num_val_files:]
        
        # Move the train files to the train folder
        for file in train_files:
            src = os.path.join(subfolder_path, file)
            sub_train_folder = os.path.join(train_folder, subfolder)
            os.makedirs(sub_train_folder, exist_ok=True)
            dst = os.path.join(sub_train_folder, file)
            shutil.move(src, dst)
        
        # Move the val files to the val folder
        for file in val_files:
            src = os.path.join(subfolder_path, file)
            sub_val_folder = os.path.join(val_folder, subfolder)
            os.makedirs(sub_val_folder, exist_ok=True)
            dst = os.path.join(sub_val_folder, file)
            shutil.move(src, dst)
        
        # Move the test files to the test folder
        for file in test_files:
            src = os.path.join(subfolder_path, file)
            sub_test_folder = os.path.join(test_folder, subfolder)
            os.makedirs(sub_test_folder, exist_ok=True)
            dst = os.path.join(sub_test_folder, file)
            shutil.move(src, dst)
```
